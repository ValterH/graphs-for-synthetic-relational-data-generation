## Rossmann

```python
from src.table_to_graph.table_to_graph import ROSSMANN_GRAPH
```

#### Visualize the subgraph belonging to node 0

```python
len(ROSSMANN_GRAPH.nodes)
```

```python
# get all descendant nodes
store_0_nodes = nx.descendants(ROSSMANN_GRAPH, 0)
store_0_nodes.add(0)

# get subgraph containing store 0 and all its descendants
store_0_subgraph = ROSSMANN_GRAPH.subgraph(store_0_nodes)
len(store_0_subgraph.nodes)
```
